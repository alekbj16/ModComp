### Introduction

The aim of this project is to gain insights into how diseases spread. In a previous project \[1], the authors used continous compartment models to study the spread of disease. Compartmend models are commonly used when modelling infectious diseases to assign a population into compartements, where each compartement describe a specific state \[2]. The model used in \[1] assumed that every individual in a population interacted with the entire population, something which is not very realistic. Further, the outcome of the simulations in \[1] where the results of model input. Compartement models are commonly solved by differential equations, which are deterministic \[3]. While compartement models depicts important charectheristics of real data, there are other methods that can be used to model the transmission of infectious diseases. The focus of the current project is to implent Monte Carlo methods. 

Monte Carlo methods are named after the Monte Carlo Casino in Monaco, because at its core it uses random numbers to solve problems \[4]. Monte Carlo methods are a class of computational algorithms which rely on random sampling to acheive numerical results \[5]. 

Mathematical modelling is of great importance when studying biological phenomenons. Models can take various forms, but the most widely used in the field of biological phenomenons are based on the random walk process \[6]. In this project, we will model how a disease spreads in a population by tracking each individual. Interactions between sick and healthy will be described by a random walk. This enables implementation and testing of various model assumptions. For a given model setup, we aim to simulate many steps forward in time, in order to compute statistics. This is the Monte Carlo component \[3]. 

In \[1], two important parameters in the compartement model approach where $\beta$, describing the effective rate of infection, and $\gamma$, describing the infection recovery rate. These two parameters gave $\mathcal{R}_0$, the basic reproduction number. However, in \[1], $\beta$ was assumed to decrease exponentially with time, and thereby ending to the pandemic. This does not accurately represent how the rate of infection has evolved over time in the Covid-19 pandemic. Approximately a year after the virus became known, there have been several waves of disease transmission. The rate of transmission have been infected by several measures, such as for example lockdown \[7]. By tracking individuals and using a random walk to describe interactions, the aim of this project is to implement and test various model assumptions. By doing so, we aim at gaining further understanding of how human behaviour affects $\beta$ in the compartement mdoel 
approach. 

### Exercise 1: Implement a random walk disease model
We start by considering a fixed population size that are confined to a finite rectangular lattice with $n_x \cdot n_y$ uniformly spaced points. Each point thus represents a specific location. During simulation, individuals are free to move at random from point to point. For each time step, every individual is assigned a certain disease status. We commence the project by assuming that there only are two possible disease states (meaning two compartements):

- Suscpetible ($S$): Healthy persons who stand the chance of becoming sick.
- Infectious ($I$): People that have the disease, and are contagious to others. 

We start the implementation of the random walk with respect to the following criterias \[3]:

- The walkers (individuals) are assumed to move randomly, meaning that there is a 25 \% chance of moving in one of the four directions: North, South, East of West. In terms of the lattice, a walker can move either in the x- or y-direction, but not in both directions at the same time. 

- At the end of a time step, after having moved each walker, check for new infections. 

- Every infectious individual will have the same probability, $q$, of infecting each susceptible person present at the same location. Go through each S-I pair in turn, and update the status of each newly infected person from $S$ to $I$.

We begin by implementing a random walk epidemic simulator class: 

(note that the class contains several functions for later exercises)

```python
# Class for Random Walk model:
class RandomWalkEpidemicSimulator:
    """
    Class used to model the spreading of a contagious disease in a
    population of individuals with a 2D random walk.

    Each walker has a disease status which is represented by an 
    integer Enum. Also, a set of integer (x, y)-coordinates are 
    stored for each walker. The possible coordinates are:

        {0, 1, ..., Lx-1} in the x-direction
        {0, 1, ..., Ly-1} in the y-direction

    It is only possible to move North, South, East, or West. If a 
    walker attempts to move outside of the physical domain, nothing 
    happens (i.e., a "bounce-back boundary condition" is enforced).
    """
    def __init__(self,
                 population_size,
                 no_init_infected=1,
                 nx=50,
                 ny=50,
                 q=0.9,
                 age=False,
                 recovery=False,
                 death = False,
                 import_inf=False,
                 vaccination=False):
        """
        :param population_size: The total number of people (N).
        :param no_init_infected: The number of infected people at t=0.
        :param nx: The number of lattice nodes in the x-direction 
        :param ny: The number of lattice nodes in the y-direction.
        :param q: The probability of infection (0 <= q <= 1).
        :param age: turn on/off age classes (if True: 50% old and young)
        :param recovery: If True, infected will recover after some time
        :param death: If True, some infected will die
        :param import_inf: If True, import infection active after min. 75 days
        :param vaccination: If True, vaccination program start after 200 days.
        """
        self.N_ = population_size
        self.I0_= no_init_infected
        self.nx_= nx
        self.ny_= ny
        self.infection_probability_ = q
        self.age_active = age
        self.recovery = recovery
        self.death = death
        self.import_inf = import_inf
        self.vaccination = vaccination
        
        # States
        self.SUSCEPTIBLE_ = 0
        self.INFECTIOUS_ = 1
        self.RECOVERED_ = 2
        self.DEAD_ = 3
        self.VACCINATED_ = 4
        
        self.Walkers_ = np.random.randint(0, 
                                  [self.nx_, self.ny_], 
                                  size=(self.N_, 2))
        self.Walkers_Old_ = self.Walkers_.copy()
        
        self.State_ = np.full(self.N_, self.SUSCEPTIBLE_) 
        self.State_[0:self.I0_] = self.INFECTIOUS_
        
        # Age class and states
        self.OLD = 0
        self.YOUNG = 1
        self.age_class = np.full(self.N_, self.YOUNG)
        indices = np.random.choice(np.arange(self.N_), replace=False, size=int(self.N_ * 0.5))
        self.age_class[indices] = self.OLD
        
        # 
        self.sickness_counter = self.State_.copy()
        self.recovery_time = np.random.normal(loc=14, scale=3, size=self.N_).astype(int)
        self.step = 0
        self.imp = False
        
        self.simulation = 0
        self.time_series = None
        
    def reset_model(self):
        """
        Resets model
        
        Returns
        -------
        None.

        """
        self.Walkers_ = np.random.randint(0, 
                                  [self.nx_, self.ny_], 
                                  size=(self.N_, 2))
        self.Walkers_Old_ = self.Walkers_.copy()
        self.State_ = np.full(self.N_, self.SUSCEPTIBLE_) 
        self.State_[0:self.I0_] = self.INFECTIOUS_
        
        self.age_class = np.full(self.N_, self.YOUNG)
        indices = np.random.choice(np.arange(self.N_), replace=False, size=int(self.N_ * 0.5))
        self.age_class[indices] = self.OLD
        
        self.sickness_counter = self.State_.copy()
        self.recovery_time = np.random.normal(loc=14, scale=3, size=self.N_).astype(int)
        
        self.step = 0
        self.imp = False  
        self.time_series
    
    def move_walkers(self):
        """
        Moves the walker one step. 
        Checks if walker move is legal (inside grid), if not, last position is kept.
        If self.age_class = True, people with state old will stay put with 20% probability
        """
        next_pos = 2*np.random.randint(0,2,size=(self.N_,2))-1
        for pi in next_pos:
            pi[np.random.randint(0,2)]=0
        # Use boolean masking instead of for loop!
        for i, walker in enumerate(self.Walkers_):
            if self.age_active ==True:
                if self.age_class[i] == self.OLD:
                    move = np.random.uniform()
                    if move <= 0.8:
                        self.Walkers_[i] += next_pos[i]
                else:
                    self.Walkers_[i] += next_pos[i]
            else:
                self.Walkers_[i] += next_pos[i]
            if (walker[0] < 0) or (walker[1] < 0) or (walker[0] > self.nx_) or (walker[1] > self.ny_):
                self.Walkers_[i] = self.Walkers_Old_[i]
        self.Walkers_Old_ = self.Walkers_.copy()
    
    def collide(self):
        """
        Checks for collision between infected and susceptible individual.
        If collision, susceptible will be infected with probability q (infection_probability_).
        If vaccination is on, a vaccinated person can be infected with probability=0.1.
        """
        # Use broadcasting to check if any walker is in same position as infectious infividual
        collision = (self.Walkers_[:,None] == self.Walkers_[self.State_ ==self.INFECTIOUS_]).all(2).any(1)
        col_index = np.array(np.where(collision)) # get index of collisions
        for i in col_index[0]:
            # If state of collided walker is susceptible, change state to infectious with 90% probability
            if self.State_[i] == self.SUSCEPTIBLE_:
                x = np.random.uniform(0,1)
                if x <= self.infection_probability_:
                    self.State_[i] = self.INFECTIOUS_
            # If state of collided walker is vaccinated, change state to infectious with 10% probability 
            if self.State_[i] == self.VACCINATED_:
                x = np.random.uniform(0,1)
                if x <= 0.1:
                    self.State_[i] = self.INFECTIOUS_      
        
    
    def plot_current_state(self):
        """
        Plot grid with current state.
        Infected: red
        Susceptible: blue
        Recovered: green
        Dead: Yellow
        Vaccinated: Purple
        """
        fig, ax = plt.subplots(1,1, figsize=(10,8))
        if (self.recovery == True) | (self.death==True):
            S=self.Walkers_[self.State_==self.SUSCEPTIBLE_]
            I=self.Walkers_[self.State_ ==self.INFECTIOUS_]
            R=self.Walkers_[self.State_==self.RECOVERED_]
            D=self.Walkers_[self.State_==self.DEAD_]
            V=self.Walkers_[self.State_==self.VACCINATED_]
            ax.scatter(I[:,0],I[:,1],c='r',s=80, label="Infected")
            ax.scatter(S[:,0],S[:,1],c='b', label="Susceptible")
            ax.scatter(R[:,0],R[:,1], c='g', label="Recovered") 
            ax.scatter(D[:,0],D[:,1], c='y',label="Dead")
            ax.scatter(V[:,0],V[:,1], c='purple',label="Vaccinated")
            ax.legend(loc=9, ncol=5, bbox_to_anchor=(0., 0.96,1.,0.102))
            ax.set_xlabel(f"No of Susceptible: {np.sum(self.State_ == self.SUSCEPTIBLE_)}\n \
No. of Infected: {np.sum(self.State_ == self.INFECTIOUS_)}\nNo. of Recovered: \
{np.sum(self.State_ == self.RECOVERED_)}\n No. of Dead: {np.sum(self.State_ == self.DEAD_)}\n \
No. of Vaccinated: {np.sum(self.State_ == self.VACCINATED_)}")
        else:
            S=self.Walkers_[self.State_==self.SUSCEPTIBLE_]
            I=self.Walkers_[self.State_ ==self.INFECTIOUS_]
            ax.scatter(I[:,0],I[:,1],c='r',s=80, label="Infected")
            ax.scatter(S[:,0],S[:,1],c='b', label="Susceptible")
            ax.legend(loc=9, ncol=2, bbox_to_anchor=(0., 0.96,1.,0.102))
            ax.set_xlabel(f"No of Susceptible: {np.sum(self.State_ == self.SUSCEPTIBLE_)}\n \
No. of Infected: {np.sum(self.State_ == self.INFECTIOUS_)}")
        ax.set_xlim(-2,self.nx_+2)
        ax.set_ylim(-2,self.ny_+2)
        ax.axhline(y=0, color='black')
        ax.axhline(y=self.ny_, color='black')
        ax.axvline(x=0, color='black')
        ax.axvline(x=self.nx_, color='black')
        ax.set_xticks(np.arange(-2,self.nx_+2,1),minor=True)
        ax.set_yticks(np.arange(-2,self.ny_+2,1),minor=True)
        ax.grid(which='minor')
        ax.set_title("Current state", y=1.06)


        
    def Beta_dt(self, t, S_tm1):
        """
        Calculates Beta*dt per time step
        
        Parameters
        ----------
        t : current time step
        S_tm1 : Susceptible at t-1 (S(t-1))

        Returns
        -------
        beta(t) : Beta at time t
        """
        I_t = np.sum(self.State_ == self.INFECTIOUS_)
        S_t = np.sum(self.State_ == self.SUSCEPTIBLE_)
        if S_t == 0:
            return float('NaN') # Returns NaN if S(t) = 0
        elif I_t == 0:
            return float('Nan') # Returns NaN if I(t) = 0
        else:
            return -((S_t-S_tm1)*self.N_)/(S_t*I_t)      

   
    def recover(self):
        """
        Called if recovery = True.
        Simulates recovery of some infected individuals. 
        The population has an individual recovery time stored in self.recovery_time. 
        This is based on a normal distribrution with mean = 14 days and sd = 3 days.
        A sickenss counter is set up to count the no. of days an individual has been infected.
        If sickness counter equals recovery time for an infected individual, the individual changes state to recovered.
        """
        # Increase sickness counter by self.State (1 for everyone in state infected, 0 for everyone in state susceptible)
        # Also increases counter for recovered, dead and vaccinated if turned on, this is then reset below.
        self.sickness_counter += self.State_
        # Sets sickness_counter for recovered, dead and vaccinated to -10 to not activate recovery.
        rec = np.where(self.State_ == self.RECOVERED_)
        self.sickness_counter[rec] = -10
        dead = np.where(self.State_ == self.DEAD_)
        self.sickness_counter[dead] = -10
        vacc = np.where(self.State_ == self.VACCINATED_)
        self.sickness_counter[vacc] = -10
        #Checks if sickness counter is equal to the individuals recovery time
        new_rec = np.where(np.equal(self.sickness_counter, self.recovery_time))[0]
        # If sickness_counter = recovery time, the individual becomes recovered
        self.State_[new_rec] = self.RECOVERED_
        
    def dead(self):
        """
        Called if death = true
        Introduces a death probability of 0.15% per day for all infected individuals.
        This corresponds to a total death probability of approx. 2.1% in 14 days.
        """
        death_prob = 0.0015
        # Array with indexes of all in state infectious
        infected_pos = np.array(np.where(self.State_ == self.INFECTIOUS_))[0]
        # Draw random number between 0 and 1 for entire population
        deaths = np.random.uniform(size=self.N_)
        # Get index where number is below death_prob
        indexes = np.array(np.where(deaths < death_prob))[0]
        # Check whether this index corresponds to and infected person
        dead_pos = np.intersect1d(indexes, infected_pos)
        # If infected person is selected, state changed to dead.
        self.State_[dead_pos] = self.DEAD_
        
        
    
    def import_infection(self):
        """
        Called if import_inf = True, time > 75 and no. of infected <= 5
        50% probability that 2 new infected will arrive in each time step. 
        Stops when no. of infected > 40 and starts again when it is 5 or below again.
        """
        # Select random number between 0 and 1 and only initiate if number is below 0.5
        rand = np.random.uniform(0,1)
        if rand < 0.5:
            new_infected = 2
            susceptible = np.sum(self.State_ == self.SUSCEPTIBLE_)
            new = min(new_infected, susceptible) #If no. of susceptible is below 'new_infected'
            sus_pos = np.array(np.where(self.State_ == self.SUSCEPTIBLE_))[0]
            # Select 'new' no. of random susceptible to change state to infectious
            indices = np.random.choice(susceptible, replace=False, size=int(new))
            inf_pos = sus_pos[indices]
            self.State_[inf_pos] = self.INFECTIOUS_

        
    def vaccination_program(self):
        """
        Called if vaccination = True and time step is > 200.
        5 people in susceptible state will change state to vaccinated each time step after step 200.
        """
        vaccinations = 5
        susceptible = np.sum(self.State_ == self.SUSCEPTIBLE_)
        # No. of vaccinated is minimum of 5 and no. of susceptible (if no. of susceptible left is less than 5)
        vacc = min(vaccinations, susceptible)
        sus_pos = np.array(np.where(self.State_ == self.SUSCEPTIBLE_))[0]
        # Choose 'vacc' no. of random people from susceptible to change state to vaccinated 
        indices = np.random.choice(susceptible, replace=False, size=int(vacc))
        vacc_pos = sus_pos[indices]
        self.State_[vacc_pos] = self.VACCINATED_        
        
        
        
    def simulate(self, steps=1):
        """
        Simulates the random walk

        Parameters
        ----------
        steps : No of steps to simulate

        Returns
        -------
        time_series : Matrix with shape [total time steps, 7]
                      M[:,0]: time
                      M[:,1]: no. of susceptible per time step
                      M[:,2]: no. of infected per time step
                      M[:,3]: beta per time step
                      M[:,4]: no. of recovered per times step
                      M[:,5]: no. of dead per time step
                      M[:,6]: no. of vaccinated per time step
        """
        if self.step == 0:
            self.time_series = np.zeros([steps, 7])
            no_recovered = np.sum(self.State_ == self.RECOVERED_)
            no_vaccinated = np.sum(self.State_ == self.VACCINATED_)
            no_susceptible = np.sum(self.State_ == self.SUSCEPTIBLE_)
            no_infectious = np.sum(self.State_ == self.INFECTIOUS_)
            no_dead = np.sum(self.State_ == self.DEAD_)
            beta = 0.06 # Chosen beta_0
            self.time_series[self.step] = [self.step, no_susceptible, no_infectious, beta, no_recovered, no_dead, no_vaccinated]
            steps = steps-1
            
        else: 
            self.time_series = np.concatenate((self.time_series, np.zeros([steps,7])))
            
        #self.simulation += 1 # Counter used when several simulations are made.
        
        for i in range(1,steps+1):
            self.step += 1
            self.collide()
            self.move_walkers()
            if self.recovery: # Active if recovery = True
                self.recover()
            if self.death: # Active if death = True
                self.dead()
            if self.imp: # Active if self.imp = True
                self.import_infection()
            if (self.step > 200) & (self.vaccination == True): # Active if vaccination = True
                self.vaccination_program()
            
            # Get no. of people in each state    
            no_recovered = np.sum(self.State_ == self.RECOVERED_)
            no_vaccinated = np.sum(self.State_ == self.VACCINATED_)
            no_susceptible = np.sum(self.State_ == self.SUSCEPTIBLE_)
            no_infectious = np.sum(self.State_ == self.INFECTIOUS_)
            no_dead = np.sum(self.State_ == self.DEAD_)
            
            # Get beta*dt per time step
            beta = self.Beta_dt(self.step, self.time_series[self.step-1,1])
            
            if self.import_inf: # Active if import_inf = True
                if (self.step > 75) & (no_infectious <= 5):
                    self.imp = True
                elif (self.step > 75) & (no_infectious > 40):
                    self.imp = False            
            
            # Save timeseries with results    
            self.time_series[self.step] = [self.step, no_susceptible, no_infectious, beta, no_recovered, 
                                           no_dead, no_vaccinated]
            
        #print(self.simulation) #Print to see simulation number.
        return self.time_series
```

We let $n_x = n_y = 50$, and set $q = 0.9$. The population size is set to $N = 683$ in order to compare with some of the work in \[1]. We suppose that there is a single infectious individual at the start of the simulation.

### Exercise 2: Comparison with deterministic SI-model

##### Part 1

This part studies the variation/uncertanty of the model.  This is done by running the simulation from the previous exercise 100 times. For each time step, we calculate the sample mean and the sample standard deviation of the number of healthy and sick people.

```python
# Make a function to run the simulator and save no. of susceptible, infected and beta in 
# separate .dat-files to not have to run the analysis every time.
def run_and_save_results(name, steps, n, population, no_init_infected=1, nx=50, ny=50, q=0.9, 
                         age=False, recovery=False, death = False, import_inf=False, vaccination=False):

    susceptible = np.zeros([n, steps])
    infected = np.zeros([n, steps])
    beta = np.zeros([n, steps])
    recovered = np.zeros([n, steps])
    dead = np.zeros([n, steps])
    vaccinated = np.zeros([n, steps])
    for i in range(n):
        run = RandomWalkEpidemicSimulator(population, no_init_infected=1, nx=50,ny=50, q=0.9, age=False, 
                                          recovery=False, death = False, import_inf=False, vaccination=False)
        timeseries = run.simulate(steps=steps)
        susceptible[i] = timeseries[:,1]
        infected[i] = timeseries[:,2]
        beta[i] = timeseries[:,3]
        recovered[i] = timeseries[:,4]
        dead[i] = timeseries[:,5]
        vaccinated[i] = timeseries[:,6]
        run.reset_model()
    np.savetxt(f"susceptible_{name}.dat", susceptible)    
    np.savetxt(f"infected_{name}.dat", infected) 
    np.savetxt(f"beta_{name}.dat", beta)
    np.savetxt(f"recovered_{name}.dat", recovered)    
    np.savetxt(f"dead_{name}.dat", dead)
    np.savetxt(f"vaccinated_{name}.dat", vaccinated)
    
    return susceptible, infected, beta, recovered, dead, vaccinated
```

```python
# run_and_save_results("p1",300,100,683)
```

In the above code, `s_matrix` and `i_matrix` contains the 100 simulations, where each of the 300 time steps have been stored. This allows for calculating the sample mean and standard deviation of number of healthy and sick people. 

```python
# Load results for steps=300 and n=100
susceptible = np.loadtxt("susceptible.dat")
infected = np.loadtxt("infected.dat")

# Get means per time step
s_means_p1 = susceptible.mean(axis=0) 
i_means_p1 = infected.mean(axis=0)

# Get standard deviation per time step
s_std_p1 = susceptible.std(axis=0) 
i_std_p1 = infected.std(axis=0)
```

```python
timeline = np.arange(0,len(s_means_p1),1)
plt.figure(figsize=(10,8))
plt.plot(timeline,s_means_p1,label="Mean number of suceptible")
plt.fill_between(timeline,s_means_p1-s_std_p1,s_means_p1+s_std_p1,alpha=0.2,color="orangered",label="Suceptible uncertanty")
plt.plot(timeline,i_means_p1,label="Mean number of infectious")
plt.fill_between(timeline,i_means_p1-i_std_p1,i_means_p1+i_std_p1,alpha=0.2,color="mediumturquoise",label="Infectious uncertanty")
plt.legend()
plt.xlabel("Time (days)")
plt.ylabel("No. of individuals")
plt.title("Expected time developement")
```

The above figure depicts the mean and uncertanty (measured by standard deviation) of the number of susceptible and infectious people in a population of 683 individuals based on a 50 $\cdot$ 50 simulation grid size. We see a clear trend in that the uncertanty shrinks at the end of the simulation period, meaning that at the end of the simulation, almost the entire population size will most likely be infected.

##### Part 3

In this part, we wish to compare the results from the two previous parts with the $SI$-model used in Project 3 \[1]. Spesifically, we wish to identify $\beta$ so that the $SI$-model approximately describes the mean number of infectious individuals as a function of time.

The analytical solution to the $SI$-model is: 

\begin{align}
I(t)=\frac{N}{1+\frac{S_0}{I_0}\exp(-\beta{t})}\,
\end{align}

where $N$ is the total population size, $S_0$ is the number of healthy people at $t=0$, $I_0$ is the number of infectious individuals at $t=0$, and $\beta(t)$ is a function that describes the effective rate of infection
per unit time, and will here be treated as constant.

```python
#Implementiing the SI-model
def I(t,N,I_0,Beta):
    S_0 = N - I_0
    return N/(1+(S_0/I_0)*np.exp(-Beta*t))

N = 683 # Population
I0_1 = 1 # Initial infected
Beta = 0.04 
max_t = 300 # Max time for plotting
dt = 1 # time step for plotting
t = np.arange(0,max_t,dt)

I_analytical_1 = I(t,N,I0_1,Beta)

#SI_model = np.empty(steps)
#for i in range(steps):
#    SI_model[i] = I(N,I_0,Beta,i)
    
```

### Exercise 3: Estimating parameters, and incorporating age classes

##### Part 1:

The original $SI$-model in Project 3 \[1] was

\begin{align}
\frac{\mathrm{d}S(t)}{\mathrm{d}t}&=-\beta(t)\cdot\frac{S(t)I(t)}{N} \,, \\ 
\frac{\mathrm{d}I(t)}{\mathrm{d}t}&=\beta(t)\cdot\frac{S(t)I(t)}{N} \,.
\end{align}

In this model, $\beta(t)$ was input to the model. However, with the random walk model we can estimate values for $\beta$ based on observed changes in the population. In order to do this, we combine the above equation for $\frac{\mathrm{d}S(t)}{\mathrm{d}t}$ with a first order approximation of the derivative \[3]. This gives the expression:

\begin{equation}
\beta\cdot\Delta t \approx -\frac{(S(t)-S(t-\Delta t))N}{S(t)I(t)}.
\end{equation}

We now use the above formula to estimate $\beta$ (note, $\Delta t$ set to 1) from estimated numbers of healthy $ S $, and infected $ I $ at each time step. This is done by estimating the mean and median of $\beta$ from several simulations. Note that a function calculating $\beta$ for each simulation has been in implemented in the class `RandomWalkEpidemicSimulator` shown in exercise 1.

### Exercise 4: Impementing our own scenario

For the final part of the project, we are going to implement new features and our own scenarios for the model. All new additions presented here are implemented in the `RandomWalkEpidemicSimulator` class given above in exercise 1. We start by introducing two new disease states: 

- Recovered ($R$): Infected individuals who have been sick for 20 days without dying becomes recovered.
- Dead ($D$): Infected individuals who unforunately die. 

**Recovered:** First, we implement the fact that infected people will recover after some amount of time. Each individual is assigned a random unique recovery time. This is drawn from a normal distribution with mean 14, represeting the expected number of days a person is contagious \[legg inn kilde], and standard deviation 3. The normal distribution was generated using `numpy.random.normal` \[https://numpy.org/doc/stable/reference/random/generated/numpy.random.normal.html]. When a individual becomes sick, a counter keeps track of how long the individual has been sick. The state of the walker is updated to recovered once the counter has reached the recovery time associated with that respective walker. This feature has been implemented in the `recover` function in the class. 

**Death:** The next step is to implement death. This is done by defining a function `death` converting some amount of infected people to dead people. We assume that the probability of dying after becoming infected is 2.2 \% \[legg inn en kilde her]. If a person does not die, then he/she will recover after some amount of time. By assuming that the total probability of dying once becoming infected is 2.2 \%, we can identify the daily death probability during the infection period. This is simply the total death probability divided by the mean number of infected days, $\frac{0.022}{14} \approx 0.0015$. For each individual walker, a random number between 0 and 1 is generated, using `numpy.random.uniform` \[https://numpy.org/doc/stable/reference/random/generated/numpy.random.uniform.html ]. The function checks for deaths in each simulation step. If a walker is infected and has a corresponding random number below the daily death probability of 0.0015, then the state of that individual is updated to infected.

**Vaccination:** After 200 days (steps), a vaccine program is introduced to the simulation. This vaccine is given to individuals who have not been infected by the disease.
